# Histogram/Distribution plot

### Histogram

```python
plt.rcParams['figure.figsize']=(18,8)

plt.subplot(1,2,1)
sns.histplot(df['total_bill'],kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],kde=True)
plt.suptitle('Univariate Analysis on Numerical Columns')
plt.show()
```

### Box plot

```python
plt.subplot(1,2,1)
sns.boxplot(df['total_bill'])

plt.subplot(1,2,2)
sns.boxplot(df['tip'])

plt.suptitle('Outliers in the Data')
plt.show()
```

### Scatter plot

```python
sns.scatterplot(x='total_bill',y='tip',data=df,hue='sex',style='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,col='size')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='sex',col='size',col_wrap=2)
plt.show()
```

### Line plot

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='line')
plt.show()
```

### Join plot

```python
sns.jointplot(x='total_bill',y='tip',data=df)
plt.show()
```

### Count plot

```python
plt.rcParams['figure.figsize']=(18,8)

plt.subplot(3,2,1)
sns.countplot(x=df['sex'])

plt.subplot(3,2,2)
sns.countplot(x=df['smoker'])

plt.subplot(3,2,3)
sns.countplot(x=df['day'])

plt.subplot(3,2,4)
sns.countplot(x=df['time'])

plt.subplot(3,2,5)
sns.countplot(x=df['size'])

plt.suptitle('Univariate Analysis on Categorical Columns')
plt.show()
```

### Bar plot

```python
sns.catplot(x='sex',y='total_bill',data=df,hue='sex')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind="bar")
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='bar',hue='size')
plt.show()
```

```python
sns.catplot(x='smoker',y='tip',data=df,kind='bar',hue='size')
plt.show()
```

### Box plot

```python
sns.boxplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",hue='smoker',data=df, kind='box')
plt.show()
```

```python
sns.catplot(x="smoker",y="total_bill",hue='sex',data=df, kind='box')
plt.show()
```
